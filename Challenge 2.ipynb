```python
im_pth = "C:\\Users\\chris\\MSOpenhack\\gear_images\\111875.jpeg"

im = Image.open(im_pth)

new_im = ImageOps.autocontrast(im, cutoff=0.5, ignore=None)
imgplot = plt.imshow(new_im)
plt.figure()
hist = new_im.histogram(mask=None, extrema=None)
plt.plot(hist)
plt.show()
```

```python

old_size = new_im.size  # old_size[0] is in (width, height) format
imgplot = plt.imshow(im)

vert_pad = max(0,old_size[0]-old_size[1])
horz_pad = max(0,old_size[1]-old_size[0])
print(vert_pad)
print(horz_pad)
padding = (horz_pad, vert_pad, 0, 0)
new_im = ImageOps.expand(new_im, padding)

imgplot = plt.imshow(new_im)
plt.figure()
hist = new_im.histogram(mask=None, extrema=None)
plt.plot(hist)
plt.show()
```

```python
new_im = new_im.resize((128,128), resample = Image.HAMMING)
imgplot = plt.imshow(new_im)
plt.figure()
hist = new_im.histogram(mask=None, extrema=None)
chans = (hist[0:256],hist[256:512],hist[512:768])
colors = ("b", "g", "r")

plt.title("'Flattened' Color Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
features = []
 
# loop over the image channels
for (chan, color) in zip(chans, colors):
    plt.plot(chan, color = color)
    plt.ylim([0,500])
    plt.xlim([0,256])

hist = hist[0:256]
plt.plot(hist)
plt.show()
```
